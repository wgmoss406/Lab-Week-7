```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x = [0.003,
0.006,
0.013,
0.026,
0.038,
0.049,
0.063,
0.074,
0.089,
0.099,
0.114,
0.126
]
y = [0.246,
0.505,
1.0144,
1.95,
3.04,
4,
5.09,
5.98,
7.17,
7.96,
9.16,
10.1
]

par = np.polyfit(x,y,1,full='True')
slope=par[0][0]

intercept=par[0][1]

variance = np.var(y)
a = [min(x), max(x)]
b = [min(x) * slope + intercept, max(x) * slope + intercept]

plt.scatter(x,y,marker='+')
plt.plot(a,b)
plt.xlabel('Current (A)')
plt.ylabel('Potential (V)')
plt.title('Resistance of an unknown resistor')
plt.show()
print('Best fit line - y = %3fx + %3f' % (slope, intercept))
```
